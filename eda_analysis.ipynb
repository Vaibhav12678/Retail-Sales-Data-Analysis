## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales_df = pd.read_csv('cleaned_reatail_sales.csv')

sales_df.head()
```

```python
sales_df.info()
```

```python
sales_df.describe()
```

```python
sales_df.isnull().sum()
```

## Total Sales

```python
total_sales = sales_df['sales'].sum()

print("Total Sales:", total_sales)
```

## Average Sales

```python
avg_sales = sales_df['sales'].mean()

print("Average Sales:", avg_sales)
```

## Sales By Category

```python
category_sales = sales_df.groupby('category')['sales'].sum()

print(category_sales)
```

## Plot Category Sales

```python
category_sales.plot(kind='bar', figsize=(8,5))

plt.title('Sales By Category')
plt.xlabel('Category')
plt.ylabel('Sales')

plt.show()
```

## Sales By Region

```python
region_sales = sales_df.groupby('region')['sales'].sum()

print(region_sales)
```

## Plot Region Sales

```python
region_sales.plot(kind='bar', figsize=(8,5))

plt.title('Sales By Region')
plt.xlabel('Region')
plt.ylabel('Sales')

plt.show()
```

```python
top_customers = sales_df.groupby('customer_name')['sales'] \
    .sum() \
    .sort_values(ascending=False) \
    .head(10)

print(top_customers)
```

## Plot Top Customers

```python
top_customers.plot(kind='bar', figsize=(10,5))

plt.title('Top 10 Customers')
plt.xlabel('Customer Name')
plt.ylabel('Sales')

plt.show()
```
